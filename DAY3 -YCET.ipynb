```python
class person:
  def __init__(self,name,age):
    self.name=name
    self.age=age

p1=person("john",43)
p2=person("Akhil",21)

print(p1.name)
print(p2.age)
```

```python
class Animal:
  def speak(self):
    print("Animals speaking")

class Dog(Animal):
  def bark(self):
    print("Dogs bark")

class Puppy(Dog):
  def eat(self):
    print("puppies eat bread")

pup = Puppy()
pup.speak()
pup.bark()
pup.eat()
```

```python
class calculation1:
  def summation(self,a,b):
    return a+b

class calculation2:
  def multiplication(self,a,b):
    return a*b

class Derived(calculation1,calculation2):
  def divide(self,a,b):
    return a/b

obj1 = Derived()
print(obj1.summation(10,20))
print(obj1.multiplication(10,20))
print(obj1.divide(10,20))
```

```python
class Bank:
  def rateofi(self):
    return 10

class sbi(Bank):
  def rateofi(self):
    return 4.5

class icici(Bank):
  def rateofi(self):
    return 5.7


b1 = Bank()
b2 = sbi()
b3 = icici()

print(f"Rate of interest for Reserve bank is {b1.rateofi()}")
print(f"Rate of interest for SBI bank is {b2.rateofi()}")
print(f"Rate of interest for ICICI bank is {b3.rateofi()}")
```

```python
class India():
    def capital(self):
        print("New Delhi is the capital of India.")

    def language(self):
        print("Hindi is the most widely spoken language of India.")

    def type(self):
        print("India is a developing country.")

class USA():
    def capital(self):
        print("Washington, D.C. is the capital of USA.")

    def language(self):
        print("English is the primary language of USA.")

    def type(self):
        print("USA is a developed country.")

obj_ind = India()
obj_usa = USA()
for country in (obj_ind, obj_usa):
    country.capital()
    country.language()
    country.type()
```
