## Question 2
Find all orders from the 'Europe' region with a total revenue greater than $500,000.

```python
mydata.filter((mydata.Region == 'Europe') & (mydata.TotalRevenue > 500000)).show()
```

## Question 3

```python
mydata.select('ItemType').filter(mydata.Region =='Middle East and North Africa').distinct().show()
```

## Question 4
Select orders that are either 'High' priority or have a total profit greater than $100,000.

```python
mydata.filter((mydata.OrderPriority =='H') |(mydata.TotalProfit > 100000)).show()
```

## Question 5
Create a new column called 'RevenueCategory' that categorizes total revenue as 'Low' if less than $100,000, 'Medium' if between $100,000 and $500,000, and 'High' if greater than $500,000.

```python
mydata.dtypes
```

```python
mydata = mydata.withColumn("TotalRevenueNew",mydata.TotalRevenue.cast("float"))
```

```python
mydata = mydata.withColumn("RevenueCategory", when(mydata.TotalRevenueNew < 100000,lit('Low')).when(((mydata.TotalRevenueNew > 500000) & (mydata.TotalRevenueNew < 100000)),lit('Medium')).otherwise(lit('High')))
```

```python
mydata.dtypes
```

```python
mydata.groupBy('RevenueCategory').agg(count('OrderId')).show()
```

## Question 7

```python
onlinedata = mydata.filter((mydata.SalesChannel == 'Online'))
onlinedata.filter((onlinedata.ItemType == 'Clothes') | (onlinedata.ItemType == 'Cosmetics')).show()
```

## Question 8
Create a new column 'ProfitMargin' that calculates the percentage of profit relative to revenue (TotalProfit / TotalRevenue * 100).

```python
mydata.withColumn('ProfitMargin',round(expr('TotalProfit/TotalRevenue*100'),2)).show()
```

## Question 9
Select distinct combinations of 'Region' and 'ItemType' where the 'OrderPriority' is 'Medium' or 'High'.

```python
mydata.select("Region","ItemType","OrderPriority").filter((mydata.OrderPriority == 'M') | (mydata.OrderPriority == 'H')).distinct().show()
```
