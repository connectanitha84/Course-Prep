```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("PythonPi").getOrCreate()
```

```python
empdata = spark.read.csv('/content/EmployeeData.csv',header=True)
empdata.show()
```

```python
orderschema = 'id int, first_name string,last_name string,email string,gender string,salary int,creationDate Date,bonus double'
df = spark.read.csv('/content/EmployeeData.csv',header=True,schema=orderschema)
df.show()
```

<b> Find the count of duplicate employee records in the input file ?

```python
df.groupBy('id').count().sort('count',ascending=False).show()
```

Find out how many records have Gender value missing.

```python
df.filter(df.gender.isNull()).count()
```

<b> Are there any missing values in the "bonus" field? If so, filled them defualt bonus 100.

```python
df.filter(df.bonus.isNull()).count()
```

```python
# from types import NoneType
# df = df.replace(to_replace=NoneType,value=100,subset=['bonus'])

df2 = df.na.fill(value=100,subset=['bonus'])
df2.show()
```

<b> Are there any employees with negative salary or bonus amounts in the input file? If so, how many?

```python
df2.filter((df2.salary < 0)| (df2.bonus < 0)).count()
```

<b> Replace all the null/emtpy value in email column with admin@azurelib.com

```python
df2.filter(df2.email.isNull()).count()
```

```python
df2 = df2.na.fill(value='admin@azurelib.com',subset=['email'])
```
