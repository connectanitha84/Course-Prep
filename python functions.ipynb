```python
number_list = [12,6,8,7,15,13,2,10,4,21]

def average_even(numbers):
  even_numbers = []
  for num in numbers:
    if num%2 == 0:
      even_numbers.append(num)

  # even_numbers = [num for num in numbers if num % 2 == 0]
  # odd_numbers = [num for num in numbers if num % 2 != 0]


  avg_even = sum(even_numbers)/len(even_numbers)
  # print(even_numbers)
  return avg_even

even_num_avg = average_even(number_list)
print(f"Average of even numbers is{even_num_avg}")
```

```python
def sum_numbers(*args):
  total = 0
  print(type(args))
  for num in args:
    total +=num
  return total

tot = sum_numbers(8,9,10)
print("Sum is ",tot)
```

```python
def sum_of_digits(number):
  if number == 0:
    return 0
  else:
    return number%10 + sum_of_digits(number // 10)


num = 12345

sum_digits = sum_of_digits(num)
print("sum is ",sum_digits)
```
