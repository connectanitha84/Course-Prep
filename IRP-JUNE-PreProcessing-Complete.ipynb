```python
data = pd.read_csv('/content/beer-servings.csv')
data.head()
```

```python
data.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
data.head()
```

```python
columns_to_change = ['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']
for col in columns_to_change:
    data[col].fillna(data[col].median(), inplace=True)

```

```python
data.isna().sum()
```

```python
import matplotlib.pyplot as plt
for col in columns_to_change:
    plt.figure()
    plt.boxplot(data[col])
    plt.title(col)
    plt.show()
```

```python
Q1 = np.percentile(data['spirit_servings'], 25, interpolation = 'midpoint')
Q2 = np.percentile(data['spirit_servings'], 50, interpolation = 'midpoint')
Q3 = np.percentile(data['spirit_servings'], 75, interpolation = 'midpoint')
```

```python
print(Q1,Q2,Q3)
```

```python
IQR = Q3- Q1
IQR
```

```python
min_val = Q1 - 1.5*IQR
max_val = Q3 + 1.5*IQR
print(min_val,max_val)
```

```python
outliers= []
for x in data['spirit_servings']:
    if((x>max_val) or (x<min_val)):
        outliers.append(x)
outliers
```

```python
data['spirit_servings'] = data['spirit_servings'].clip(min_val,max_val)
```

```python
Q1 = np.percentile(data['wine_servings'], 25, interpolation = 'midpoint')
Q2 = np.percentile(data['wine_servings'], 50, interpolation = 'midpoint')
Q3 = np.percentile(data['wine_servings'], 75, interpolation = 'midpoint')
IQR = Q3- Q1
min_val = Q1 - 1.5*IQR
max_val = Q3 + 1.5*IQR
print(min_val,max_val)
data['wine_servings'] = data['wine_servings'].clip(min_val,max_val)
```
