```python
import numpy as np
```

```python
speed = [90,98,88,90,110,103,87,90,78,85,86]
speed_mean = np.mean(speed)
print(speed_mean)
```

```python
speed_median = np.median(speed)
print("Median is ",speed_median)
```

```python
from scipy.stats import mode #mode cannot be found out using Numpy Library
speed_mode = mode(speed)
speed_mode
```

```python
print("mode is ",speed_mode[0])
```

```python
from statistics import median, mode

new_median = median(speed)
new_mode = mode(speed)

print("Median : ",new_median)
print("Mode : ",new_mode)
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 95],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female'],
    'Salary': [50000, 60000, 75000, 80000, 90000],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami'],
    'Score1': [25,72,65,72,82],
    'Score2':[85,10,56,98,87]
}
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```
