```python
import random as random

students = ["Jane","Joe","Beck","Tom","Liam","Culhane",
            "Candance","Ana","Mari","Steffi","Adam"]

marks=[]
for i in range(0,len(students)):
     marks.append(random.randint(0, 101))

plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("CLASS RECORDS")
plt.xticks(rotation=45)
plt.plot(students, marks, color = 'k',
         linestyle = 'dotted', marker = '*',
         markerfacecolor = 'c', markersize = 8, markeredgecolor = 'r')
```

```python
# Creating dataset
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']
nums = [23, 17, 35, 29, 12, 41]
explode=[0.1, 0, 0, 0,0.1,0]

# Creating plot
fig = plt.figure(figsize =(10, 4))
plt.pie(nums, labels = cars,explode=explode , autopct ="%1.1f%%")
plt.title("Car sold")

# show plot
plt.show()
```

```python
# first data point
x = [1, 2, 3, 4]
y = [4, 1, 3, 6]

# depict first scatter plot
plt.scatter(x, y, c='green')

# second data point
x1 = [5, 6, 7, 8]
y1 = [1, 3, 5, 2]

# depict second scatter plot
plt.scatter(x1, y1, c='red',marker="s")
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [4, 1, 3, 6]

x1 = [5, 6, 7, 8]
y1 = [1, 3, 5, 2]

fig,ax=plt.subplots(1,2,sharey=True)
ax[0].plot(x,y)
ax[0].set_title('Line chart ')

ax[1].scatter(x1,y1)
ax[1].set_title('scatter chart')
```
