## Treating the outlier by Imputing

```python
q1 = hd['Sale Price'].quantile(0.25)
q3 = hd['Sale Price'].quantile(0.75)

iqr = q3 - q1
```

```python
iqr
```

```python
upper_limit = q3 + 1.5*iqr
lower_limit = q1 - 1.5*iqr
```

```python
print(upper_limit)
print(lower_limit)
```

```python
def limit_imputer(value):
    if value>upper_limit:
        return upper_limit
    if value<lower_limit:
        return lower_limit
    else:
        return value
```

```python
# applying the imputer function
hd['Sale Price'] = hd['Sale Price'].apply(limit_imputer)
```

## Encoding

```python
mapping = {'None':'0',
          'Once':'1',
          'Twice':'2',
          'Thrice':'3',
          'Four':'4'}

hd['No of Times Visited'] = hd['No of Times Visited'].map(mapping)
```

```python
# etracting information from renovated year

hd['Ever Renovated']=np.where(hd['Renovated Year']==0,'No','Yes')
```

```python
# extracting Year from Date of House sold

hd['Purchase Year'] = pd.DatetimeIndex(hd['Date House was Sold']).year
```

```python
hd['Years since renovation'] = np.where(hd['Ever Renovated']=='Yes', abs(hd['Purchase Year']-hd['Renovated Year']),0)
```
