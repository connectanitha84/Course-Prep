MATPLOTLIB & SEABORN

```python
df_viz = pd.read_excel('/content/medals.xls')
df_viz.head()
```

```python
df_viz = df_viz.rename(columns={'Unnamed: 0':'Country'})
df_viz.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.bar(df_viz['Country'],df_viz['Gold'])
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Gold Medals')
plt.title('Country & Gold Medals')
plt.show()
```

```python
plt.boxplot(df_viz['Gold'])
plt.show()
```

```python
sns.scatterplot(x='Gold',y='Silver', data=df_viz)
plt.show()
```

```python
sns.histplot(df_viz['Gold'])
plt.show()
```

```python
sns.pairplot(df_viz)
plt.show()
```

Missing Value Handling

```python
dict = {'First Score':[100, 90, np.nan, 95,np.nan],
        'Second Score': [30, 45, 56, np.nan,np.nan],
        'Third Score':[np.nan, 40, 80, 98,np.nan],
        'Fourth Score':[np.nan, np.nan, np.nan, 65,np.nan],
        'Fifth Score':[12,15,18,43,54]}
```

```python
df = pd.DataFrame(dict)
df
```

```python
df.isna().sum()
```

```python
df.fillna(10)
```

```python
df.mean()
```

```python
df.fillna(df.mean())
```

```python
df.fillna(method='ffill')
```

```python
replacevals = {'First Score':0,'Second Score':1,'Third Score':2,'Fourth Score':3,'Fifth Score':4}
df.fillna(value=replacevals)
```

```python
df.dropna()
```

```python
df.dropna(how='all')
```

```python
df.dropna(axis=1)
```
