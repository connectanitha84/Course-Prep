```python
# Keyword Arguments
# The idea is to allow the caller to specify the argument name with values so that the caller does not need to remember the order of parameters.

def greet_person(name, age, location="Unknown"):
    """A function that greets a person with their name, age, and location."""
    print(f"Hello {name}! You are {age} years old and from {location}.")

# Example function calls with keyword parameters
greet_person(name="Alice", age=30, location="New York")
greet_person(name="Bob", age=25)  # The location parameter will use the default value

# In this example, we have a function called greet_person() that takes three parameters: name, age, and location.
# The location parameter has a default value of "Unknown" assigned to it.

# When we call the function greet_person, we pass the arguments as keyword parameters.

# In the first function call, we provide all three arguments explicitly.
# In the second function call, we only provide name and age, and the location parameter uses its default value.
```

```python
# Arbitrary Keyword Arguments

def sum_numbers(*args):
    total = 0
    for num in args:
        total += num
    return total

result = sum_numbers(1, 2, 3, 4, 5)
print("Sum:", result)
```

```python
def sum_numbers(*args):
    print(type(args))
    result = 0
    for num in args:
        result += num
    return result

# Call the function with various numbers of arguments
print(sum_numbers(1, 2, 3))       # Output: 6
print(sum_numbers(10, 20, 30, 40)) # Output: 100
print(sum_numbers(5, 7))          # Output: 12
```

```python
# key-value arguemtns

def print_info(**kwargs):
    print(type(kwargs))
    for key, value in kwargs.items():
        print(f"{key}: {value}")

# Call the function with various keyword arguments
print_info(name="Alice", age=30, city="New York")
print_info(country="USA", occupation="Engineer")
```

```python
def greet_person(**kwargs):
    """A function that greets a person with their name, age, and location."""
    name = kwargs.get('name', 'Unknown')
    age = kwargs.get('age', 'Unknown')
    location = kwargs.get('location', 'Unknown')

    print(f"Hello {name}! You are {age} years old and from {location}.")

# Call the function using key-value pair arguments
greet_person(name="Alice", age=30, location="New York")
greet_person(name="Bob", age=25)
greet_person(location="London", name="Charlie", age=22)
```

```python
def change_variable_with_argument(x):
    x = 10 #local variable

x = 5
print("Before function call, x =", x)  # Output: Before function call, x = 5

change_variable_with_argument(x)
print("After function call, x =", x)

# the first method using the global keyword modifies the global variable directly, while the second method using an argument only changes
# the local variable inside the function without affecting the original variable outside the function.
```

```python
def change_variable():
   global x #global variable
   x = 10

x = 5
print("Before function call, x =", x)  # Output: Before function call, x = 5
change_variable()
print("After function call, x =", x)
```

```python
""" Global Scope: Variables defined outside any function or class have a global scope, making them accessible throughout the entire program.

Local Scope: Variables defined inside a function have a local scope, making them accessible only within that specific function. """

global_var = 10  # Global variable

def example_function():
    local_var = 5  # Local variable
    #global global_var  # Declare that we want to use the global variable inside the function
    global_var = 100
    print("Inside the function:")
    print("Local variable:", local_var)
    print("Global variable:", global_var)

print("\nOutside the function:")
print("Global variable (before modification):", global_var)
# Call the function
example_function()

# Modify the global variable outside the function
#global_var = 20

print("Global variable (after modification):", global_var)
```

```python
# Example for zip():

names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
cities = ["New York", "San Francisco", "London"]

# Using zip() to combine multiple lists
zipped_data = zip(names, ages, cities)
# zipped_data = list(zip(names, ages, cities))

print(zipped_data)

# Iterating over the zipped data
for name, age, city in zipped_data:
    print(f"{name} is {age} years old and lives in {city}")
```

```python
zipped_data = [('Alice', 25, 'New York'), ('Bob', 30, 'San Francisco'), ('Charlie', 35, 'London')]

# Using zip() with * operator to unzip the data
names, ages, cities = zip(*zipped_data)

print("Names:", names)
print("Ages:", ages)
print("Cities:", cities)
```
