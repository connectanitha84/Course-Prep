```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
```

```python
data.head()
```

```python
data.drop('User ID', axis=1, inplace=True)
```

```python
X = data.drop('Purchased', axis=1)
y = data['Purchased']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.25,random_state=42)
```

```python
#SVM model
from sklearn.svm import SVC
svmclf = SVC(kernel='linear')
svmclf.fit(X_train, y_train)
```

```python
y_pred = svmclf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print('Accuracy is: ', accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
#Decision tree
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train, y_train)
y_pred = dt_clf.predict(X_test)
```

```python
print('Accuracy is: ', accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
#Random forest
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier()
rf_clf.fit(X_train, y_train)
```

```python
y_pred = rf_clf.predict(X_test)
```

```python
print('Accuracy is: ', accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
#Hyper parameter tuning
rf_clf1 = RandomForestClassifier(n_estimators=20, max_depth=20, criterion='entropy', min_samples_split=3)
rf_clf1.fit(X_train, y_train)


# n_estimators= The required number of trees in the Random Forest. The default value is 10. We can choose any number but need to take care of the overfitting issue.
# criterion= It is a function to analyze the accuracy of the split. Here we have taken "entropy" for the information gain.
```

```python
y_pred = rf_clf1.predict(X_test)
```

```python
print('Accuracy is: ', accuracy_score(y_test, y_pred))
```
