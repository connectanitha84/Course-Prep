```python
import random
randomno=[]
for i in range(4):
  n=random.randint(0,9)
  randomno.append(n)
print(randomno)
tries=0
while(tries<10): # checking if the number of tries is less than 10 or not
  tries+=1 #incrementing the number of tries
  guess=input("Enter the four digit code:- ")
  guessno= list(map(int, guess)) #converting entered input value to a list of integer values
  if(len(guess)!=4): # checking the length of entered string
    print("Please enter 4 digit number")
    continue;
  if(randomno == guessno): #checking if the auto created value and user entered are similar
    print("Congratulations! You have guessed the right number.")
    break;
  else:
    for index, value in enumerate(guessno):# itereating through the guessed number
      if (value in randomno)  : #checking if each value is present in the auto created list
        if value == int(randomno[index]): #check if the value is in the same position as it is created
          print("R",end="")
        else:
          print("Y",end="")
      else:
        print("N",end="")
    print("\n")


  
 
```
