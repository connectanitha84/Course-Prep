```python
words = ["chair","table","mouse","laptop","keyboard"]
vowels = "aeiou"
count = 0
for i in words: #iterating the list elements
  print("list data is : ",i)
  for j in i: #iterating the characters in each word
    if j in vowels: #checking if the character in vowels
      print(j)
      count += 1
  print("Count is : ",count)
  count = 0
```

```python
words = ["Malayalam","chair","table","mouse","laptop","keyboard"]
vowels = "aeiou"
vowelsinword =[]
count = 0
for i in words: #iterating the list elements
  print("list data is : ",i)
  for j in i: #iterating the characters in each word
    if j in vowels: #checking if the character in vowels string
      if j not in vowelsinword:
        vowelsinword.append(j)
  print("Count is : ",len(vowelsinword))
  print("Vowels are : ",vowelsinword)
  vowelsinword =[]
  count = 0
```

```python
students = [
    {"name": "Alice", "age": 24, "grades": [85, 90, 92]},
    {"name": "Bob", "age": 22, "grades": [78, 81, 86]},
    {"name": "Charlie", "age": 23, "grades": [92, 89, 85]},
    {"name": "David", "age": 21, "grades": [75, 70, 72]},
    {"name": "Eva", "age": 24, "grades": [88, 94, 90]}
]

# Initialize variables to track the student with the maximum grade
max_grade = -1
top_student = None
top_student_Total = None
max_total_grade = -1

# Calculate the average grade for each student and find the student with the highest grade
for student in students:
    average_grade = sum(student["grades"]) / len(student["grades"])
    student["average_grade"] = average_grade
    print(f"{student['name']}'s average grade is {average_grade:.2f}")

    # Check for the highest individual grade
    student_max_grade = max(student["grades"])
    if student_max_grade > max_grade:
        max_grade = student_max_grade
        top_student = student["name"]

    total_grade = sum(student["grades"])
    student["total_grade"] = total_grade
    print(f"{student['name']}'s total grade is {total_grade}")

    # Check for the highest total grade
    if total_grade > max_total_grade:
        max_total_grade = total_grade
        top_student_Total = student["name"]

# Print the student with the highest individual grade
print(f"The student with the highest individual grade is {top_student} with a grade of {max_grade}.")
```
