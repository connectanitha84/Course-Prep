Extracting features and labels

````python
labels = df['DEATH_EVENT']
features = df.drop(columns=['DEATH_EVENT'])
features
````

````python
labels
````

````python

# c1-50
# c2-100
# c1-50 c2-50  #undersampling
# c1-100 c2-100 #oversampling
#X - features
#y - labels
X = features
y = labels
````

````python
# Display class distribution before addressing imbalance
print("Class Distribution Before Sampling:")
class_counts = y.value_counts()
print(class_counts)
````

````python
#Imbalance  C0 - 203, C1-96
X_imbalanced = X
y_imbalanced = y
# Apply Random Over-sampling
#c0 - majority 203
#c1 - minority 96 +107
#over
#c0-203
#c1-203
from imblearn.over_sampling import RandomOverSampler
# Apply Random Over-sampling
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X_imbalanced, y_imbalanced)
# Display class distribution after oversampling
print("\nClass Distribution After Random Over-sampling:")
print(y_resampled.value_counts())
````

````python
from imblearn.under_sampling import RandomUnderSampler
# Apply Random Under-sampling
rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X_imbalanced, y_imbalanced)
#under
#c0-96
#c1-96
# Display class distribution after undersampling
print("\nClass Distribution After Random Under-sampling:")
print(y_resampled.value_counts())
````

````python
from imblearn.over_sampling import SMOTE
# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_imbalanced, y_imbalanced)
# Display class distribution after SMOTE
print("\nClass Distribution After SMOTE:")
print(y_resampled.value_counts())
````

###Feature Selection Techniques

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#Correlation based feature selection

# Concatenate X and y along columns (axis=1)
df = pd.concat([X_resampled, y_resampled], axis=1)
# Calculate correlation matrix
corr_matrix = df.corr()


# Plot correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Matrix')
plt.show()

#-1 to 1
#col1-col2 corr=.8 highly correlated
#col1-col2 corr=-0.6 not correlated
#col1-col2 corr=0.05 not correlated
````

````python
#1. Correlation based feature selection
import pandas as pd
import numpy as np
# Calculate correlations between features and target variable
correlations = X_resampled.corrwith(y_resampled)
print(correlations)
# Absolute correlations
correlations = correlations.abs()

# Sort correlations in descending order
correlations = correlations.sort_values(ascending=False)
print(correlations)

# Select the top n most correlated features
n = 5  # Adjust the number of features you want to select
selected_features = correlations.head(n).index.tolist()
print("Selected features based on correlation with target variable:")
print(selected_features)
````

````python

# Plot correlations
plt.figure(figsize=(10, 6))
correlations.plot(kind='bar')
plt.title('Correlation with Target Variable')
plt.xlabel('Features')
plt.ylabel('Correlation')
plt.xticks(rotation=45)
plt.show()
````

````python
#SelectKBest method
#scoring function ANOVA F-Statistic
from sklearn.feature_selection import SelectKBest, f_classif
# Create SelectKBest object with ANOVA F-statistic scoring function
#applied to continuous or categorical features
k_best = SelectKBest(score_func=f_classif, k=5)  # Select the top 5 features
X_k_best = k_best.fit_transform(X_resampled, y_resampled)
# Print the selected features
selected_features = k_best.get_support(indices=True)
print("Selected Features:", selected_features)
````

````python
from sklearn.feature_selection import SelectKBest, chi2
#chi2
# applied to categorical features to assess their relationship with a categorical target variable.
# SelectKBest with chi-squared
k_best_chi2 = SelectKBest(score_func=chi2, k=5)
X_chi2 = k_best_chi2.fit_transform(X_resampled, y_resampled)
print("Selected Features (Chi-squared):")
print("Number of Selected Features:", X_chi2.shape[1])
print("List of Selected Features:", k_best_chi2.get_support(indices=True))
print("\n")
````

````python
#mutual info
from sklearn.feature_selection import SelectKBest, mutual_info_classif
# Apply SelectKBest with mutual information
# feature selection based on mutual information for classification tasks
#For classification tasks, mutual_info_classif calculates the mutual information between each feature and the target variable.
k_best_mi = SelectKBest(score_func=mutual_info_classif, k=5)  # Adjust 'k' based on your preference
X_mi = k_best_mi.fit_transform(X_resampled, y_resampled)
print("List of Selected Features:", k_best_mi.get_support(indices=True))
````

````python
from sklearn.feature_selection import VarianceThreshold
# Apply VarianceThreshold
threshold = 0.3  # Adjust the threshold based on your preference
variance_selector = VarianceThreshold(threshold=threshold)
X_variance = variance_selector.fit_transform(X_resampled)
X_variance

# Get the computed threshold values
threshold_values = variance_selector.variances_
print(threshold_values)
print("List of Selected Features:", variance_selector.get_support(indices=True))
# Get the selected features
selected_features = X_resampled.columns[variance_selector.get_support()]

# Show selected features
print("Selected Features:")
print(selected_features)
````

###Dimensionality Reduction

````python
#PCA
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
# Apply PCA to reduce to 3 dimensions
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_resampled)
X_pca.shape
X_pca
````

````python
# Plot the reduced data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_resampled, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: Disease Predition')
plt.show()
````

````python
#TSNE
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Apply t-SNE to reduce to 3 dimensions
tsne = TSNE(n_components=3)
X_tsne = tsne.fit_transform(X_resampled)
print(X_tsne.shape)
X_tsne
# Plot the reduced data
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y_resampled, cmap='viridis')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.title('t-SNE: Iris Dataset')
plt.show()
````
