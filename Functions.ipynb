```python
# Sample list of numbers
number_list = [23, 56, 12, 67, 45, 78, 33, 90, 24, 17]

# Implement the function to calculate the average of even numbers
def calculate_average_even(numbers):
    even_numbers = [num for num in numbers if num % 2 == 0]
    if even_numbers:
        return sum(even_numbers) / len(even_numbers)
    else:
        return 0

# Implement the function to calculate the average of odd numbers
def calculate_average_odd(numbers):
    odd_numbers = [num for num in numbers if num % 2 != 0]
    if odd_numbers:
        return sum(odd_numbers) / len(odd_numbers)
    else:
        return 0

# Display the results
average_even = calculate_average_even(number_list)
average_odd = calculate_average_odd(number_list)

print(f"The average of even numbers is: {average_even:.2f}")
print(f"The average of odd numbers is: {average_odd:.2f}")
```

You are tasked with creating a program that analyzes a list of numbers. Write a Python function named analyze_numbers that takes a list of integers as input and returns a dictionary containing the following information:

The sum of all numbers in the list.
The average (mean) of the numbers.
The maximum and minimum numbers in the list.
Implement the function and demonstrate its usage by calling it with a sample list of numbers.

```python
def analyze_numbers(numbers):
    if not numbers:
        return None  # Return None for an empty list

    result = {
        'sum': sum(numbers),
        'average': sum(numbers) / len(numbers),
        'maximum': max(numbers),
        'minimum': min(numbers)
    }
    return result

# Sample list of numbers
number_list = [5, 12, 8, 20, 15]

# Call the function and display the results
result_dict = analyze_numbers(number_list)

if result_dict is not None:
    print(f"List: {number_list}")
    print(f"Sum: {result_dict['sum']}")
    print(f"Average: {result_dict['average']:.2f}")
    print(f"Maximum: {result_dict['maximum']}")
    print(f"Minimum: {result_dict['minimum']}")
else:
    print("The list is empty.")
```

```python
numbers = [1, 2, 3, 4, 5]

squared_numbers = list(map(lambda x: x * x, numbers))

print("Original numbers:", numbers)
print("Squares of numbers:", squared_numbers)

#We start by defining a list called numbers containing some integer values: [1, 2, 3, 4, 5].

# We then use the map() function to apply a given function (in this case, a lambda function) to each item in the iterable (in this case, the list numbers).
# The map() function returns an iterator that yields the results.

# The lambda function lambda x: x * x is defined to take a single argument x and return the square of x.
# It is an anonymous function, meaning it doesn't have a name and is defined in place.

# The map() function applies this lambda function to each element of the numbers list.
# For example, it first applies the lambda function to 1, resulting in 1 * 1 = 1. Then it applies the lambda function to 2, resulting in 2 * 2 = 4, and so on.

# The result of the map() function is an iterator. To see the actual values, we convert this iterator to a list using list(), and
# store it in the variable squared_numbers.

# Finally, we print out both the original list of numbers (numbers) and the list containing the squares of those numbers (squared_numbers).
```

```python
#recursive example function

def sum_of_digits(n):
    if n == 0:
        return 0
    else:
        return n % 10 + sum_of_digits(n // 10)

# Example usage:
number = 12345
print("Sum of digits of", number, "is", sum_of_digits(number))
```

```python
# Example for zip():

names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
cities = ["New York", "San Francisco", "London"]

# Using zip() to combine multiple lists
zipped_data = zip(names, ages, cities)
# zipped_data = list(zip(names, ages, cities))

# print(list(zipped_data))

# Iterating over the zipped data
for name, age, city in zipped_data:
    print(f"{name} is {age} years old and lives in {city}")


# The zip() function in Python is used to combine multiple iterables (such as lists, tuples, or other sequences) element-wise into tuples.
# It takes any number of iterables as input and returns an iterator of tuples where the i-th tuple contains the i-th element from each of
# the input iterables.
```

```python
zipped_data = [('Alice', 25, 'New York'), ('Bob', 30, 'San Francisco'), ('Charlie', 35, 'London')]

# Using zip() with * operator to unzip the data
names, ages, cities = zip(*zipped_data)

print("Names:", names)
print("Ages:", ages)
print("Cities:", cities)
```

Implement the following functions to manage a user's account:

create_account(accounts, account_number, initial_balance): This function should create a new account with the given account number and initial balance. It should update the accounts dictionary with the new account information.

deposit(accounts, account_number, amount): This function should deposit the specified amount into the account with the given account number. If the account does not exist, print an error message.

withdraw(accounts, account_number, amount): This function should withdraw the specified amount from the account with the given account number. If the account does not exist or if the balance is insufficient, print an error message.

check_balance(accounts, account_number): This function should display the current balance of the account with the given account number. If the account does not exist, print an error message.

```python
# Full code
accounts = {}

def create_account(accounts, account_number, initial_balance):
    if account_number not in accounts:
        accounts[account_number] = initial_balance
        print(f"Account {account_number} created with an initial balance of {initial_balance}.")
    else:
        print(f"Error: Account {account_number} already exists.")

def deposit(accounts, account_number, amount):
    if account_number in accounts:
        accounts[account_number] += amount
        print(f"Deposited {amount} into account {account_number}.")
    else:
        print(f"Error: Account {account_number} does not exist.")

def withdraw(accounts, account_number, amount):
    if account_number in accounts:
        if accounts[account_number] >= amount:
            accounts[account_number] -= amount
            print(f"Withdrew {amount} from account {account_number}.")
        else:
            print(f"Error: Insufficient balance in account {account_number}.")
    else:
        print(f"Error: Account {account_number} does not exist.")

def check_balance(accounts, account_number):
    if account_number in accounts:
        print(f"The balance in account {account_number} is {accounts[account_number]}.")
    else:
        print(f"Error: Account {account_number} does not exist.")

# Example usage
create_account(accounts, "A123", 1000)
deposit(accounts, "A123", 500)
withdraw(accounts, "A123", 200)
check_balance(accounts, "A123")
create_account(accounts, "A163", 1000)
deposit(accounts, "A163", 500)
withdraw(accounts, "A163", 200)
check_balance(accounts, "A163")
accounts
```
