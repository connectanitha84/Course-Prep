```python
df = pd.read_csv('/content/beer-servings.csv')
df.head()
```

```python
df['beer_servings'] = df['beer_servings'].fillna(df['beer_servings'].median())
df['spirit_servings'] = df['spirit_servings'].fillna(df['spirit_servings'].median())
df['wine_servings'] = df['wine_servings'].fillna(df['wine_servings'].median())
df['total_litres_of_pure_alcohol'] = df['total_litres_of_pure_alcohol'].fillna(df['total_litres_of_pure_alcohol'].median())

df.isna().sum()
```

```python
Q1 = np.percentile(df['spirit_servings'],25,method='midpoint')
Q2 = np.percentile(df['spirit_servings'],50,method='midpoint')
Q3 = np.percentile(df['spirit_servings'],75,method='midpoint')
```

```python
print("Q1 is",Q1)
print("Q2 is",Q2)
print("Q3 is",Q3)
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
low_lim = Q1 - 1.5*IQR
up_lim = Q3 + 1.5*IQR

print(low_lim)
```

```python
print(up_lim)
```

```python
outlier =[]
for x in df['spirit_servings']:
  if((x <low_lim) or (x> up_lim)):
    outlier.append(x)

outlier
```

```python
max(df['spirit_servings'])
```

```python
df['spirit_servings'] =df['spirit_servings'].clip(lower =low_lim , upper = up_lim)
```

```python
from scipy.stats import zscore
z_scores = zscore(df['wine_servings'])
print(z_scores)
```

```python
# Define a threshold (e.g., 3 standard deviations)
threshold = 3

# Find indices of outliers
outlier_indices = df.index[(z_scores > threshold) | (z_scores < -threshold)]

print(outlier_indices)
# Remove rows with outliers using the indices
df_no_outliers = df.drop(outlier_indices)

# Display the DataFrame without outliers
print("DataFrame without Outliers:")
print(df_no_outliers)
```
