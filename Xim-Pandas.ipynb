## DATAFRAME

```python
data = {'Year':[2001,2002,2004,2005,2006],
        'States':['Kerala','TN','Andhra', 'Karnataka', 'Bihar'],
        'Pop':[300,250,280,400,360]}

pop_data = pd.DataFrame(data)
pop_data
```

```python
pop_data = pd.DataFrame(pop_data , columns=['Year','States','Pop','Income'])
pop_data
```

```python
pop_data['Income'] =[20000,30000,15000,23000,60000]
pop_data
```

```python
pop_data['States']
```

```python
pop_data['Debt'] = [320,430,235,540,230]
pop_data
```

```python
pop_data['expense'] = 750
pop_data
```

```python
pop_data['savings'] = pop_data['Income'] - pop_data['Debt']
pop_data
```

```python
del pop_data['Pop']
```

```python
pop_data
```
