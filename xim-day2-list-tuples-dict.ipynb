```python
#Sort() & Reverse()

mylist = ['apple', 'banana', 'cherry', 'orange', 'kiwi', 'melon', 'mango', 'tomato']
```

```python
mylist.reverse() #reverse the elements in the list
mylist
```

```python
mylist.sort() #sort elements in ascending order
print(mylist)
```

```python
mylist.sort(reverse=True) #descending order
print(mylist)
```

```python
#Count() & len()
# vowels list
vowels = ['a', 'e', 'i', 'o', 'i', ['d','f','g']]

# count element 'i'
count = vowels.count('i')

# print count
print('The count of i is:', count)

length = len(vowels)
print("number of elements is :",length)
```
