```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
data = pd.read_csv('/content/Social_Network_Ads.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data1 = data.drop('User ID',axis=1)
data1.columns
```

```python
data1 =pd.get_dummies(data1)
data1.head()
```

```python
data1['Purchased'].value_counts()
```

```python
X= data1.drop('Purchased',axis=1)
y= data1['Purchased']

X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state =42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LRmodel = LogisticRegression()
LRmodel = LRmodel.fit(X_train,y_train)
```

```python
LR_score = LRmodel.score(X_test,y_test)
LR_score
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
```

```python
rf_model = rf_model.fit(X_train,y_train)
RF_score = rf_model.score(X_test,y_test)
RF_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
KN_model = KNeighborsClassifier()
KN_model = KN_model.fit(X_train,y_train)
KN_score = KN_model.score(X_test,y_test)
KN_score
```

```python
# KFOLD TESTING
```

```python
from sklearn.model_selection import KFold
kfold_validator  = KFold(10)
```

```python
for train_index, test_index in kfold_validator.split(X,y):
  print('Training Index: ', train_index)
  print('Test Index: ', test_index)
```

```python
from sklearn.model_selection import cross_val_score
cv_result = cross_val_score(rf_model,X,y,cv=kfold_validator)
cv_result
```
