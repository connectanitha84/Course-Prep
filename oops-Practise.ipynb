Class Methods vs. Static Methods vs. Instance Methods


```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Instance method
    def introduce(self):
        return f"My name is {self.name}, and I am {self.age} years old."

    # Class method
    @classmethod
    def from_birth_year(cls, name, birth_year):
        age = cls.calculate_age(birth_year)
        return cls(name, age)

    @staticmethod
    def calculate_age(birth_year):
        current_year = 2023  # Assuming the current year is 2023
        return current_year - birth_year

# Creating instances using instance method
person1 = Person("Alice", 25)
person2 = Person("Bob", 30)

# Creating an instance using a class method
person3 = Person.from_birth_year("Eve", 1990)

# Using static method
age = Person.calculate_age(1985)

# Using instance methods
print(person1.introduce())
print(person2.introduce())

# Using the instance method on the object created by the class method
print(person3.introduce())

# Using the static method
print(f"Bob's age is {age} years.")

# Instance Methods: introduce is an instance method. It operates on the instance of the class and can access and
# modify instance-specific data (in this case, name and age).

# Class Method: from_birth_year is a class method. It takes the class itself (cls) as its first argument and is often used for
# creating instances with specific behaviors.
#  In this case, it calculates the age based on birth year and creates a Person object.Static Method: calculate_age is a static method.
#  It does not depend on the class or instance's state and behaves like a regular function.
#   It can be called on the class itself without creating an instance.
```

Polymorphism is a fundamental concept in object-oriented programming that allows objects of different classes to be treated as objects of a common superclass. This allows you to write more generic code that can work with objects of various types. In Python, polymorphism can be achieved through method overriding, where subclasses provide their own implementation of a method that is already defined in their superclass.

The polymorphism comes into play when we iterate through the list of shapes and call the area method on each shape. The appropriate area method is invoked based on the actual object type. This demonstrates how you can work with objects of different types through a common interface, thanks to polymorphism.

```python
class Shape:
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159265359 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Create instances of Circle and Rectangle
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Calculate and print the areas of different shapes
shapes = [circle, rectangle]

for shape in shapes:
    print(f"Area of the shape: {shape.area()}")
```

**Encapsulation** is one of the fundamental principles of object-oriented programming (OOP) that involves bundling data (attributes) and the methods (functions) that operate on that data into a single unit, known as a class. The idea is to hide the internal details and protect the integrity of an object by controlling access to its attributes, which are often marked as private and can only be accessed and modified through the object's methods. Encapsulation helps maintain the internal state of an object and ensures that it's used in a consistent and controlled manner.

In this example:

We have a Person class with two private attributes, __name and __age, which are not meant to be accessed directly from outside the class. These attributes are encapsulated and can only be accessed and modified through the class's methods.

The class provides getter and setter methods (get_name, set_name, get_age, and set_age) to access and modify the private attributes. The setter methods include validation to ensure that names are at least 2 characters long and ages are non-negative.

We create a person object and demonstrate how to get and set the attributes using the provided methods. This way, we encapsulate the data and provide controlled access to it.

```python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def get_name(self):
        return self.__name

    def set_name(self, name):
        if len(name) > 1:
            self.__name = name
        else:
            print("Name must be at least 2 characters long.")

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age >= 0:
            self.__age = age
        else:
            print("Age must be a non-negative number.")

# Create a person object
person = Person("Alice", 30)

# Access and modify attributes using methods
print("Initial Name:", person.get_name())
person.set_name("Bob")
print("Updated Name:", person.get_name())

print("Initial Age:", person.get_age())
person.set_age(-5)  # This will not change the age
print("Updated Age:", person.get_age())
```

Inheritance is a fundamental concept in object-oriented programming (OOP) that allows you to create a new class (called a subclass or derived class) based on an existing class (called a superclass or base class). Inheritance allows the subclass to inherit attributes and methods from the superclass, promoting code reuse and the creation of a hierarchy of classes.

```python
# Superclass (Base Class)
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print(f"The {self.year} {self.make} {self.model}'s engine is started.")

    def stop_engine(self):
        print(f"The {self.year} {self.make} {self.model}'s engine is stopped.")

# Subclass 1
class Car(Vehicle):
    def __init__(self, make, model, year, fuel_type):
        super().__init__(make, model, year)
        self.fuel_type = fuel_type

    def honk_horn(self):
        print(f"The {self.year} {self.make} {self.model} honks the horn: Beep! Beep!")

# Subclass 2
class Bicycle(Vehicle):
    def __init__(self, make, model, year, gear_count):
        super().__init__(make, model, year)
        self.gear_count = gear_count

    def ring_bell(self):
        print(f"The {self.year} {self.make} {self.model} rings the bell: Ding! Ding!")

# Create instances of Car and Bicycle
my_car = Car("Toyota", "Camry", 2022, "Gasoline")
my_bicycle = Bicycle("Schwinn", "Mountain Bike", 2021, 21)

# Use methods from both the superclass and subclass
my_car.start_engine()
my_car.honk_horn()

my_bicycle.start_engine()
my_bicycle.ring_bell()
```

**Data Abstraction** is a concept in object-oriented programming (OOP) that focuses on hiding the complex implementation details of an object while exposing a simple and high-level interface for interacting with that object. In other words, data abstraction allows you to emphasize what an object does rather than how it does it. It simplifies the interaction with objects and helps manage complexity in large software systems.

We have a BankAccount class that represents a simple bank account. It has attributes such as account_number, account_holder, and balance.

The class provides methods like deposit, withdraw, and get_balance for interacting with the bank account. These methods abstract the complexity of managing the account's balance and provide a high-level interface for depositing and withdrawing money.

Users of the BankAccount class don't need to know the inner workings of how the account balance is managed; they can interact with the account using the methods provided. This is data abstraction in action, as it hides the implementation details while exposing a simplified way to work with the object.

Data abstraction helps in managing complexity, improving code maintainability, and reducing the risk of errors by providing a clear and consistent interface to interact with objects. It allows you to focus on what an object does rather than the low-level details of how it achieves its functionality.

```python
class BankAccount:
    def __init__(self, account_number, account_holder):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = 0

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount

    def withdraw(self, amount):
        if amount > 0 and self.balance >= amount:
            self.balance -= amount
        else:
            print("Insufficient balance or invalid amount.")

    def get_balance(self):
        return self.balance

# Create a BankAccount instance
account = BankAccount("123456", "John Doe")

# Deposit and withdraw money using the provided methods
account.deposit(1000)
account.withdraw(500)

# Get the account balance using the get_balance method
balance = account.get_balance()
print(f"Account balance: ${balance}")
```

```python
from abc import ABC, abstractmethod

# Abstract superclass defining the interface
class BankAccount(ABC):
    def __init__(self, account_number, account_holder):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = 0

    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

    def get_balance(self):
        return f"Account Number: {self.account_number}, Holder: {self.account_holder}, Balance: ${self.balance}"

# Concrete subclass implementing the abstract methods
class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder, interest_rate):
        super().__init__(account_number, account_holder)
        self.interest_rate = interest_rate

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
        else:
            print("Insufficient funds!")

# Concrete subclass implementing the abstract methods
class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder, overdraft_limit):
        super().__init__(account_number, account_holder)
        self.overdraft_limit = overdraft_limit

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if self.balance + self.overdraft_limit >= amount:
            self.balance -= amount
        else:
            print("Insufficient funds!")

# Create instances of concrete subclasses
savings_account = SavingsAccount("SA123456", "Alice", 0.03)
checking_account = CheckingAccount("CA789012", "Bob", 1000)

# Use the abstraction to deposit, withdraw, and get balances
savings_account.deposit(1000)
checking_account.deposit(500)

savings_account.withdraw(200)
checking_account.withdraw(1500)

print(savings_account.get_balance())
print(checking_account.get_balance())
```

```python
from abc import ABC, abstractmethod

class MyAbstractClass(ABC):
    @abstractmethod
    def my_abstract_method(self):
        pass

class MyConcreteClass(MyAbstractClass):
    def my_abstract_method(self):
        print("Concrete implementation of my_abstract_method")

# Attempting to create an instance of the abstract class would result in a TypeError:
#my_instance = MyAbstractClass()  # This will raise a TypeError

# Creating an instance of the concrete subclass is allowed:
my_instance = MyConcreteClass()
my_instance.my_abstract_method()  # This will call the concrete implementation
```
