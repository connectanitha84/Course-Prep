```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 33]

# Create a figure with multiple subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 8)) # 2 rows and 2 columns

# Bar chart
axs[0, 0].bar(categories, values) #first figure is at the (0,0) position .ie Ist row Ist column
axs[0, 0].set_title('Bar Chart')

# Line chart
axs[0, 1].plot(categories, values, marker='o')  #second figure is at the (0,1) position .ie Ist row 2nd column
axs[0, 1].set_title('Line Chart')

# Pie chart
axs[1, 0].pie(values, labels=categories, autopct='%1.1f%%') #third figure is at the (1,0) position .ie 2nd row 1st column
axs[1, 0].set_title('Pie Chart')

# Scatter plot
axs[1, 1].scatter(categories, values) #fourth figure is at the (1,1) position .ie 2nd row 2nd column
axs[1, 1].set_title('Scatter Plot')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```
