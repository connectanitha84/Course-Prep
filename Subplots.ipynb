```python
import matplotlib.pyplot as plt
import numpy as np
# Create a figure and a grid of subplots with 2 rows and 3 columns
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 6))

# You can access each subplot using the 'axes' array
# For example, to plot on the top-left subplot:
axes[0, 0].plot([0, 1, 2], [0, 1, 4], label='Line 1')
axes[0, 0].set_title('Subplot 1')

axes[0, 1].plot([0, 1, 2], [1, 2,3], label='Line 1')
axes[0, 1].set_title('Subplot 2')

axes[0, 2].scatter([0, 1, 2], [1, 2,3],color='green', label='Line 1')
axes[0, 2].set_title('Subplot 3')

axes[1, 0].plot([0, 1, 2], [1, 2,3], label='Line 1')
axes[1, 0].set_title('Subplot 4')

axes[1, 1].scatter([0, 1, 2], [1, 2,3], label='Line 1')
axes[1, 1].set_title('Subplot 5')

axes[1, 2].scatter([0, 1, 2], [0, 1, 4], color='red', label='Points')
axes[1, 2].set_title('Subplot 6')

# For example, setting common labels for x and y axes:
fig.text(0.5, 0.04, 'X-axis label', ha='center', va='center')
fig.text(0.06, 0.5, 'Y-axis label', ha='center', va='center', rotation='vertical')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Create a figure and a grid of subplots with 2 rows and 3 columns
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))

# Plotting on the first subplot (top-left)
axes[0, 0].plot([0, 1, 2], [0, 1, 4], label='Line Plot')
axes[0, 0].set_title('Line Plot')

# Plotting on the second subplot (top-center)
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
axes[0, 1].plot(x, y, label='Sine Wave')
axes[0, 1].set_title('Sine Wave')

# Plotting on the third subplot (top-right)
data = {'A': 20, 'B': 45, 'C': 30, 'D': 5}
axes[0, 2].bar(data.keys(), data.values(), color='skyblue')
axes[0, 2].set_title('Bar Chart')

# Plotting on the fourth subplot (bottom-left)
axes[1, 0].scatter(np.random.rand(30), np.random.rand(30), color='orange', label='Random Scatter')
axes[1, 0].set_title('Scatter Plot')

# Plotting on the fifth subplot (bottom-center)
axes[1, 1].hist(np.random.randn(1000), bins=30, color='green', alpha=0.7)
axes[1, 1].set_title('Histogram')

# Plotting on the sixth subplot (bottom-right)
labels = 'Apples', 'Bananas', 'Cherries', 'Grapes'
sizes = [30, 25, 20, 25]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
axes[1, 2].pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
axes[1, 2].set_title('Pie Chart')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
