Question: You have been observing the frequency of visits by different bird species at a bird feeder during a 24-hour period. You hypothesize that the visits are evenly distributed among four bird species: Sparrows, Robins, Blue Jays, and Cardinals. You want to test whether the observed frequencies match the expected frequencies based on your hypothesis.

You have collected the following data:

Observed Frequencies:
Frequency of visits by bird species at a bird feeder during a 24-hour period
Observed Frequencies:

Sparrows: 30 visits
Robins: 20 visits
Blue Jays: 25 visits
Cardinals: 25 visits

```python
Ho ="bird visits are evenly distributed"
H1 ="visits are not evenly distributed"
```

```python
import numpy as np
from scipy.stats import chisquare

alpha =0.05
observed = np.array([30,20,25,25])
expected = np.array([25,25,25,25])

chi2 , p = chisquare(observed, f_exp = expected)

print("Chi-square statistics : ",chi2)
print("p-value : ",p)
```

```python
# p-value  < alpha

# reject the null hypothesis

# p-value > alpha

# we fail to reject the null hypothesis

```

Question: Is there a significant association between a person's political affiliation and their preferred mode of transportation to work?

PoliticalAffiliation: ['Democrat', 'Republican', 'Independent', 'Democrat', 'Independent', 'Republican']

TransportationMode: ['Car', 'Public Transit', 'Bike', 'Car', 'Car', 'Public Transit']

    


```python
Ho = "There is no association between political affiliation and transportation mode"
H1 = "There is association between the two"

from scipy.stats import chi2_contingency
import pandas as pd

data = pd.DataFrame({
    'PoliticalAffiliation': ['Democrat', 'Republican', 'Independent', 'Democrat', 'Independent', 'Republican'],
    'TransportationMode': ['Car', 'Public Transit', 'Bike', 'Car', 'Car', 'Public Transit']})

contingency_table = pd.crosstab(data['PoliticalAffiliation'],data['TransportationMode'])

contingency_table
```

```python
chi2 , p_val , dof , expected = chi2_contingency(contingency_table)
print("Chi-square value = ",chi2)
print("P value = ",p_val)
print("Degree of freedom = ",dof)
print("Expected frequency = ",expected)
```

```python
alpha =0.05
```

To determine whether to accept or reject the null hypothesis, you can compare the p-value to the significance level as follows:

If the p-value is less than or equal to the significance level (p â‰¤ alpha), you would reject the null hypothesis.
If the p-value is greater than the significance level (p > alpha), you would fail to reject the null hypothesis.
In your case, the p-value is 0.09, which is greater than the significance level of 0.05. Therefore, you would fail to reject the null hypothesis. This means that there is not enough evidence to conclude that there is a significant association between political affiliation and transportation mode in the dataset you analyzed. The variables appear to be independent or not strongly associated.
