```python
# Conditional statement to determine the grade based on a score
score = int(input("Enter your total score"))

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print(f"Your grade is: {grade}")
```

```python
# Input: Weather condition
weather = input("What's the weather like today? (sunny/rainy/cloudy): ")

# Conditional statement to decide what to wear
if weather == "sunny":
    print("It's sunny! Wear sunglasses, a hat, and sunscreen.")
elif weather == "rainy":
    print("It's rainy! Grab your umbrella and wear a waterproof jacket.")
elif weather == "cloudy":
    print("It's cloudy! You may want a light jacket just in case.")
else:
    print("I'm not sure about that weather condition. Check a weather app for advice.")
```

```python
# Simulating a traffic light
traffic_light_color = "Green"

if traffic_light_color == "Red":
    print("Stop! Do not proceed. Wait for the green light.")

elif traffic_light_color == "Yellow":
    print("Slow down and prepare to stop if it's safe.")

elif traffic_light_color == "Green":
    print("Go ahead. The road is clear.")

else:
    print("The traffic light is malfunctioning. Proceed with caution.")
```

```python
# Real-world scenario: Admission eligibility based on age and previous education

# Input: Age and previous education level
age = int(input("Enter your age: "))
education = input("Enter your highest education level (high school/college): ")

# Conditional statement to check eligibility
if age >= 18 and education == "college":
    print("You are eligible for university admission.")
elif age >= 18 and education == "high school":
    print("You are eligible for college admission.")
elif age < 18 and education == "high school":
    print("You are not yet eligible for college, but you can pursue higher education after high school.")
else:
    print("Sorry, you are not eligible for university or college admission at this time.")
```

```python
# Real-world scenario: Eligibility for a store discount
# Input: Age and membership status
age = int(input("Enter your age: "))
is_member = input("Are you a store member? (yes/no): ").lower()

# Conditional statement using if, elif, and else with 'and' and 'or' operators
if age >= 60 and is_member == "yes":
    print("Congratulations! You qualify for a senior member discount.")
elif age >= 60 or is_member == "yes":
    print("You are eligible for a store discount.")
else:
    print("Sorry, you do not qualify for any discounts at the moment.")
```

```python
# Grocery list
grocery_list = ["apples", "bananas", "bread", "milk", "eggs"]

# Item to search for
item_to_find = "pen"

# Use a for loop to search for the item in the list
for item in grocery_list:
    if item == item_to_find:
        print(f"{item_to_find} is on the list.")
        break
else:
    print(f"{item_to_find} is not on the list.")
```

```python
#WHILE LOOPS

# Define the savings goal and initial savings
savings_goal = 1000
current_savings = 0

# Define the monthly deposit amount
monthly_deposit = 250

# Initialize a variable to count the number of months
months = 0

# Use a while loop to track savings progress
while current_savings < savings_goal:
    current_savings += monthly_deposit
    months += 1

# Print the result
print(f"Congratulations! You reached your savings goal of ${savings_goal} in {months} months.")
```
