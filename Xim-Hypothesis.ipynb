A coffee shop claims that their coffee is served at a mean temperature of 170 degrees Fahrenheit. You suspect that the coffee is not as hot as they claim. You collect a sample of 20 cups of coffee and measure their temperatures. The sample mean temperature is 167 degrees Fahrenheit with a standard deviation of 3 degrees. Perform a hypothesis test at a 5% significance level to determine whether the coffee is served at a temperature different from 170 degrees Fahrenheit.


```python
import scipy.stats as stats

H0 = "Coffee served at 170 degree"
H1 = "Coffee not served at 170 degree"

pop_mean  = 170
sample_mean = 167
sample_std = 3
sample_size = 20
alpha = 0.05

# Calculate the standard error of the mean
standard_error = pop_mean / (sample_size ** 0.5)

# Calculate the t-statistic
t_statistic = (sample_mean - pop_mean) / standard_error

p_value = stats.t.cdf(t_statistic, df=sample_size-1)

if p_value < alpha:
    print("Reject the null hypothesis (HO)")
else:
    print("Fail to reject the null hypothesis (H1)")


print("p_value is : ",p_value)
```

#CHI-SQUARE TEST


Question: You have been observing the frequency of visits by different bird species at a bird feeder during a 24-hour period. You hypothesize that the visits are evenly distributed among four bird species: Sparrows, Robins, Blue Jays, and Cardinals. You want to test whether the observed frequencies match the expected frequencies based on your hypothesis.

You have collected the following data:

Observed Frequencies:
Frequency of visits by bird species at a bird feeder during a 24-hour period
Observed Frequencies:
Sparrows: 30 visits
Robins: 20 visits
Blue Jays: 25 visits
Cardinals: 25 visits

Your expected distribution (null hypothesis) is that each bird species visits the feeder equally (25% for each). You want to determine whether there is a significant difference between the observed and expected frequencies of bird visit


```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies
observed = np.array([30, 20, 25, 25])

# Expected frequencies based on the null hypothesis (even distribution)
expected = np.array([25, 25, 25, 25])

# Perform the goodness-of-fit chi-square test
chi2, p = chisquare(observed, f_exp=expected)

# Print the results
print("Chi-Square Statistic:", chi2)
print("P-value:", p)
```

In this case, the null hypothesis is that the visits by different bird species are evenly distributed, and the alternative hypothesis is that they are not evenly distributed. The chi-square test will help you determine whether the observed frequencies significantly differ from the expected frequencies, indicating whether the birds have a preference for certain times of the day.

 If the p-value for the goodness-of-fit chi-square test is 0.5724 (rounded to four decimal places), the result would typically be interpreted as follows:

Since the p-value is relatively high (greater than the common significance level of 0.05 or 5%), you would fail to reject the null hypothesis. In this context, it means that there is not enough evidence to conclude that the observed bird visit frequencies significantly differ from the expected frequencies based on an even distribution among the bird species.

In other words, the data does not provide sufficient support to suggest that the birds have a strong preference for specific times of the day at the bird feeder. The observed and expected frequencies are consistent with the hypothesis of an even distribution.

Question: Is there a significant association between a person's political affiliation and their preferred mode of transportation to work?

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
data = pd.DataFrame({
    'PoliticalAffiliation': ['Democrat', 'Republican', 'Independent', 'Democrat', 'Independent', 'Republican'],
    'TransportationMode': ['Car', 'Public Transit', 'Bike', 'Car', 'Car', 'Public Transit']
})

contingency_table = pd.crosstab(data['PoliticalAffiliation'], data['TransportationMode'])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-Square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)
```

In this example, the null hypothesis is that there is no association between political affiliation and transportation mode, and the alternative hypothesis is that they are associated. The p-value will help you determine if there is a significant association.

You can use the chi2_contingency function from the SciPy library to perform the chi-square test and obtain the test statistic, p-value, degrees of freedom, and expected frequencies. You can then interpret the results to determine whether there is a significant association between the variables.

Question 2: Goodness of Fit

You are a wildlife biologist studying the distribution of bird species at a specific location in a forest. Based on your prior research, you hypothesize that four bird species (A, B, C, D) should be equally distributed in the area. You collect data from a sample of 200 bird sightings and observe the following distribution:

Species A: 55 sightings
Species B: 45 sightings
Species C: 50 sightings
Species D: 50 sightings
Perform a chi-square goodness-of-fit test using Python to determine whether the observed distribution matches the expected distribution based on your hypothesis. Use a significance level of 0.05.

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies
observed = np.array([55, 45, 50, 50])

# Expected frequencies based on the null hypothesis (even distribution)
expected = np.array([50, 50, 50, 50])

# Perform the goodness-of-fit chi-square test
chi2, p = chisquare(observed, f_exp=expected)

# Significance level (alpha)
alpha = 0.05

# Print the results and make a decision
print("Chi-Square Statistic:", chi2)
print("P-value:", p)

if p <= alpha:
    print("Reject the null hypothesis. The observed distribution is significantly different from the expected distribution.")
else:
    print("Fail to reject the null hypothesis. The observed distribution matches the expected distribution.")
```
